**2.1 Obtendo o percentual de funcion√°rios com notas abaixo de 7 por n√≠vel hier√°rquico**

Come√ßaremos contando a quantidade de avalia√ß√µes abaixo de 7 em cada uma das habilidades comportamentais de acordo com o n√≠vel hier√°rquico correspondente.

Em seguida, dividiremos esse total pela quantidade total de funcion√°rios do respectivo n√≠vel hier√°rquico para cada habilidade. O resultado ser√° multiplicado por 100 para obter o percentual correspondente.

Por fim, reuniremos todos esses percentuais em uma lista.

Veja esse processo sendo aplicado com os estagi√°rios da empresa:

```python
percentual_estagiarios=[]
nivel=df[df['N√≠vel']=='Estagi√°rio'][['Trabalho em Equipe','Lideran√ßa','Comunica√ß√£o','Iniciativa','Organiza√ß√£o']]
qtd_estagiarios=len(nivel)

#Percentual de est√°giarios com pontua√ß√£o menor que 7 no quesito trabalho em equipe
percentual_estagiarios.append((nivel[nivel['Trabalho em Equipe']<7]['Trabalho em Equipe'].count()/qtd_estagiarios*100).round(2))

#Percentual de estagi√°rios com pontua√ß√£o menor que 7 no quesito lideran√ßa
percentual_estagiarios.append((nivel[nivel['Lideran√ßa']<7]['Lideran√ßa'].count()/qtd_estagiarios*100).round(2))

#Percentual de estagi√°rios com pontua√ß√£o menor que 7 no quesito comunica√ß√£o
percentual_estagiarios.append((nivel[nivel['Comunica√ß√£o']<7]['Comunica√ß√£o'].count()/qtd_estagiarios*100).round(2))

#Percentual de estagi√°rios com pontua√ß√£o menor que 7 no quesito iniciativa
percentual_estagiarios.append((nivel[nivel['Iniciativa']<7]['Iniciativa'].count()/qtd_estagiarios*100).round(2))

#Percentual de estagi√°rios com pontua√ß√£o menor que 7 no quesito organiza√ß√£o
percentual_estagiarios.append((nivel[nivel['Organiza√ß√£o']<7]['Organiza√ß√£o'].count()/qtd_estagiarios*100).round(2))

#Lista contendo os percentual de avalia√ß√µes abaixo de 7 dos estagi√°rios em cada habilidade comportamental
percentual_estagiarios
```
